```python
X=df.drop('number',axis=1)
y=df['number']
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=100,stratify=y)
lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.decomposition import PCA
pc=PCA()
```

```python
X_pca=pc.fit_transform(X)
```

```python
pca_df=pd.DataFrame(columns=['column name','explained variance','cummulative sum'])
```

```python
pca_df['column name']=range(0,64)
```

```python
pca_df['explained variance']=pc.explained_variance_ratio_
```

```python
pca_df['cummulative sum']=pca_df['explained variance'].cumsum()
```

```python
pca_df[pca_df['cummulative sum']>0.94]
```

```python
pc=PCA(n_components=40)
X_pca=pc.fit_transform(X)
```
